```python
import pybamm
import matplotlib.pyplot as plt
import os

os.chdir(pybamm.__path__[0] + "/..")
```

#### which model

```python
model_DFN = pybamm.lithium_ion.DFN()
model_SPM = pybamm.lithium_ion.SPM()
model_SPMe = pybamm.lithium_ion.SPMe()

# O'Regan 2022 parameter set
param = pybamm.ParameterValues("OKane2022")
param.set_initial_stoichiometries(1);
```

```python
s = pybamm.step.string
# s("Rest for 1 hour", period="30 minutes"),
experiment = pybamm.Experiment(
    [
        (
            "Discharge at 1 C until 2.5 V",
            "Rest for 1 hour",
            "Charge at 4 C until 4.2 V",
            "Hold at 4.2 V until C/50",
            "Rest for 1 hour",
        )
    ] * 50,
    termination="80% capacity",
)
```

```python
sols, sims = [], []
for model in [model_DFN, model_SPM, model_SPMe]:
    globals()[f'sim_{model}'] = pybamm.Simulation(model, experiment=experiment, parameter_values=param)
    globals()[f'sol_{model}'] = globals()[f'sim_{model}'].solve(calc_esoh=True, showprogress=True)
    sims.append(globals()[f'sim_{model}'])
    sols.append(globals()[f'sol_{model}'])
```

```python
for sol in sols:
    print(sol.solve_time)
```

```python
# sol.plot(['Current [A]', 'Voltage [V]'])
pybamm.plot_summary_variables(solutions=sols,
                              labels=['DFN', 'SPM', 'SPMe'],
                              output_variables=['Capacity [A.h]', 
                               'Change in loss of capacity to negative lithium plating [A.h]',
                               'Change in loss of capacity to positive lithium plating [A.h]',
                               'Change in loss of active material in negative electrode [%]',
                               'Change in loss of active material in positive electrode [%]',
                               'Change in loss of capacity to negative SEI [A.h]',
                               'Change in loss of capacity to positive SEI [A.h]',
                               'Change in loss of capacity to negative SEI on cracks [A.h]',
                               'Change in loss of capacity to positive SEI on cracks [A.h]'
                              ])
```
